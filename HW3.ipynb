### Problem 1

```python
x = sp.Symbol('x')
L = sp.Symbol('L')
n = sp.Symbol('n')
sp.nsimplify((2 / (L)) * sp.integrate(x**2 *
                                      sp.sin(n * x * sp.pi / L), (x, 0, L)))
```

```python
f = lambda x: x * x
L = 1
```

```python
N = 5
p1 = FourierSeries(f,L,series_type='sin')
p1.series(N)
```

```python
p1.b[3]
```

```python
min_coef = 1
max_coef = 10
p1.plot_f_series_range(min_coef=1,max_coef=5).show()
```
